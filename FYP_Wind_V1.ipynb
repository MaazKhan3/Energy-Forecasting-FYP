```python
import numpy as np

def calculate_swept_area(diameter):
    """
    Calculates the swept area of the turbine's rotor.
    """
    radius = diameter / 2
    swept_area = np.pi * (radius ** 2)
    return swept_area

def calculate_daily_energy(turbine_factors, power_curve, daily_wind_speeds):
    """
    Calculates the average wind speed, swept area, and daily energy production for a wind turbine.

    Parameters:
        turbine_factors (list): A list containing [Rated_Power, Cutin_Speed, Cutout_Speed, Rated_Speed, RotorDiameter, AVG_DailyWindSpeed]
        power_curve (list): A list where each index represents the power output for wind speeds from 1 m/s up to 15 m/s.
        daily_wind_speeds (list): List of daily wind speeds in m/s.

    Returns:
        dict: Dictionary containing average wind speed, swept area, and total daily energy produced (kWh).
    """
    # Unpacking turbine factors
    rated_power = turbine_factors[0]
    cutin_speed = turbine_factors[1]
    cutout_speed = turbine_factors[2]
    rated_speed = turbine_factors[3]
    rotor_diameter = turbine_factors[4]

    # Calculate swept area
    swept_area = calculate_swept_area(rotor_diameter)

    # Calculate average daily wind speed
    #avg_wind_speed = np.mean(daily_wind_speeds)
    avg_wind_speed = 6.34

    # Calculate daily energy production
    total_energy = 0  # in kWh
    for wind_speed in daily_wind_speeds:
        if wind_speed < cutin_speed or wind_speed >= cutout_speed:
            # No energy produced outside operating wind speeds
            power_output = 0
        elif cutin_speed <= wind_speed < rated_speed:
            # Interpolate power from the power curve based on wind speed
            power_index = int(wind_speed) - 1  # Adjusting index to match curve (1 m/s at index 0)
            power_output = power_curve[power_index]
        else:
            # At rated wind speed and above, within operating range, use rated power
            power_output = rated_power

        # Convert power (kW) to energy (kWh) over 1 hour for simplicity
        total_energy += power_output

    return {
        "Average_Wind_Speed": avg_wind_speed,
        "Swept_Area": swept_area,
        "Total_Daily_Energy": total_energy  # Total energy in kWh
    }

# Example usage
turbine_factors = [225, 4, 25, 13, 29, None]  # Example values for Rated_Power, Cutin_Speed, Cutout_Speed, Rated_Speed, RotorDiameter, AVG_DailyWindSpeed
power_curve = [0.0, 0.0, 0.0, 7.10, 20.50, 38.30, 61.90, 92.20, 127.90, 164.90, 196.40, 216.00, 223.0, 225.0, 225.0]  # Example power curve (kW)
daily_wind_speeds = [6.34, 6.55, 6.70, 6.80, 6.91, 7.11, 7.30, 7.45, 7.54, 7.58, 7.66, 7.58, 7.66, 7.57, 7.46, 7.44, 7.48, 7.53, 7.41, 7.17, 6.83, 5.37, 5.55, 5.71, 5.87, 6.03, 6.26]  # Example daily wind speeds in m/s

# Calculate energy and swept area
result = calculate_daily_energy(turbine_factors, power_curve, daily_wind_speeds)

print("Average Wind Speed:", result["Average_Wind_Speed"], "m/s")
print("Swept Area:", result["Swept_Area"], "m^2")
print("Total Daily Energy Produced:", result["Total_Daily_Energy"], "kWh")
```

```python
import requests

def get_wind_speed(city_name, api_key):
    """
    Fetches real-time wind speed for a specified city using the Weatherstack API.

    Parameters:
        city_name (str): Name of the city for which wind speed is to be fetched.
        api_key (str): Your API key for Weatherstack.

    Returns:
        float: The wind speed in kilometers per hour, or None if an error occurs.
    """
    url = f"http://api.weatherstack.com/current?access_key={api_key}&query={city_name}"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for unsuccessful requests
        data = response.json()

        # Check for success and extract wind speed
        if "current" in data:
            wind_speed = data['current']['wind_speed']
            print(f"Wind speed in {city_name}: {wind_speed} km/h")
            return wind_speed
        else:
            print(f"Error fetching data: {data.get('error', {}).get('info', 'Unknown error')}")
            return None

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

# Example usage:
api_key = "2a090ce181d80b96535cb1c98bda649c"  # Replace with your Weatherstack API key
city_name = "London"
get_wind_speed(city_name, api_key)
```

```python
import requests
from datetime import datetime, timedelta

def get_daily_average_wind_speed(city_name, api_key):
    """
    Fetches the average wind speed for the current day for a specified city
    using the Weatherstack API.

    Parameters:
        city_name (str): Name of the city for which wind speed is to be fetched.
        api_key (str): Your API key for Weatherstack.

    Returns:
        float: The average wind speed in kilometers per hour, or None if an error occurs.
    """
    # Get today's date
    today = datetime.now()
    today_str = today.strftime('%Y-%m-%d')  # Format date as YYYY-MM-DD

    # Construct URL for historical data
    url = f"http://api.weatherstack.com/historical?access_key={api_key}&query={city_name}&historical_date={today_str}"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for unsuccessful requests
        data = response.json()

        # Check for success and extract wind speed data
        if "historical" in data:
            historical_data = data['historical'][today_str]
            hourly_data = historical_data['hourly']

            # Extract wind speeds from hourly data
            wind_speeds = [hour['wind_speed'] for hour in hourly_data]
            average_wind_speed = sum(wind_speeds) / len(wind_speeds) if wind_speeds else 0

            print(f"Average wind speed in {city_name} for today: {average_wind_speed:.2f} km/h")
            return average_wind_speed
        else:
            print(f"Error fetching data: {data.get('error', {}).get('info', 'Unknown error')}")
            return None

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

# Example usage:
api_key = "2a090ce181d80b96535cb1c98bda649c"  # Replace with your Weatherstack API key
city_name = "London"
get_daily_average_wind_speed(city_name, api_key)
```

```python
import requests
import pandas as pd

# Define function to fetch wind speed data from NASA POWER API
def fetch_wind_speed(lat, lon, start_year, end_year):
    """
    Fetch hourly wind speed data from NASA POWER API for a specific location and date range.

    Args:
        lat (float): Latitude of the location.
        lon (float): Longitude of the location.
        start_year (int): Start year (e.g., 2016).
        end_year (int): End year (e.g., 2020).

    Returns:
        pd.DataFrame: DataFrame with wind speed data.
    """
    base_url = "https://power.larc.nasa.gov/api/temporal/hourly/point"
    params = {
        "parameters": "WS10M",  # Wind speed at 10 meters
        "community": "RE",
        "longitude": lon,
        "latitude": lat,
        "start": f"{start_year}0101",
        "end": f"{end_year}1231",
        "format": "JSON",
    }

    response = requests.get(base_url, params=params)

    if response.status_code != 200:
        raise Exception(f"API request failed with status code {response.status_code}: {response.text}")

    data = response.json()["properties"]["parameter"]["WS10M"]

    # Convert JSON data to DataFrame
    wind_speed_df = pd.DataFrame(list(data.items()), columns=["DateTime", "Wind Speed (m/s)"])
    wind_speed_df["DateTime"] = pd.to_datetime(wind_speed_df["DateTime"], format="%Y%m%d%H")

    return wind_speed_df

# Define location (Karachi coordinates) and time range
latitude = 24.8607
longitude = 67.0011
start_year = 2016
end_year = 2020

# Fetch wind speed data
try:
    wind_speed_data = fetch_wind_speed(latitude, longitude, start_year, end_year)
    print(wind_speed_data.head())
    # Save data to CSV
    wind_speed_data.to_csv("karachi_wind_speed.csv", index=False)
    print("Wind speed data saved to 'karachi_wind_speed.csv'.")
except Exception as e:
    print(f"Error: {e}")
```

```python
# Function to calculate swept area
def calculate_swept_area(diameter):
    radius = diameter / 2
    return np.pi * (radius ** 2)

# Function to calculate daily energy production
def calculate_daily_energy(avg_wind_speed, turbine_factors, power_curve):
    rated_power, cutin_speed, cutout_speed, rated_speed, rotor_diameter = turbine_factors

    # Calculate swept area
    swept_area = calculate_swept_area(rotor_diameter)

    # If wind speed is outside the turbine's operational range
    if avg_wind_speed < cutin_speed or avg_wind_speed >= cutout_speed:
        return 0.0

    # Calculate power output
    if cutin_speed <= avg_wind_speed < rated_speed:
        # Interpolate power from the power curve
        power_index = int(avg_wind_speed) - 1  # Index matches wind speed (1 m/s at index 0)
        power_output = power_curve[power_index]
    else:
        # Use rated power within operational range
        power_output = rated_power

    # Convert power (kW) to energy (kWh) over 24 hours
    return power_output * 24

# Create a list to store daily energy for each turbine
daily_energy_data = {}

# Iterate through each turbine
for idx, turbine in TD.iterrows():
    turbine_id = turbine["Turbine_ID"]
    turbine_factors = [
        turbine["Rated_Power_kW"],
        turbine["Cut_in_Speed_m/s"],
        turbine["Cut_out_Speed_m/s"],
        turbine["Rated_Speed_m/s"],
        turbine["Rotor_Diameter_m"],
    ]
    power_curve = turbine["Power_Curve_kW"]

    # Calculate daily energy for this turbine
    daily_energy = []
    for _, row in daily_wind_speeds.iterrows():
        avg_wind_speed = row["Avg_Wind_Speed_m/s"]
        energy = calculate_daily_energy(avg_wind_speed, turbine_factors, power_curve)
        daily_energy.append(energy)

    # Store the daily energy list for this turbine
    daily_energy_data[turbine_id] = daily_energy

# Print the daily energy data for each turbine
for turbine_id, energy_list in daily_energy_data.items():
    print(f"Turbine {turbine_id} Daily Energy (kWh):")
    print(energy_list)
    print()
```
